### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np

random_array = np.random.randint(1, 51, size=10)
media = np.mean(random_array)
media2 = np.average(random_array)
desvio_padrao = np.std(random_array)

print("Array criado: ", random_array)
print("Media calculada com mean: ", media)
print("Media calculada com average: ", media2)
print("Desvio padrao calculado: ", desvio_padrao)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
import numpy as np
array = np.random.rand(10)
array[array <= 0.5] = np.sqrt(array[array <= 0.5])

print("Array criado: ", array)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
import numpy as np
# Cria uma matriz 3x3x3 com valores aleatorios
matriz = np.random.rand(3,3,3)

# Exibe o valor maximo de cada linha da matriz 2D dentro da matriz 3D criada
valores_maximos = np.amax(matriz, axis=1)

# Exibe os resultados
print("Matriz 3D criada:\n", matriz, "\n")
print("Valores maximos de cada matriz 2D:\n", valores_maximos)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
import numpy as np
# Cria uma matriz de tamanho 5x5
matriz = np.random.rand(5,5)

# Subtrair a media de cada linha da matriz
media = np.mean(matriz, axis=1)
matriz_linha_media = matriz - media[:, np.newaxis]

print("Matriz original:\n", matriz,"\n")
print("Media de cada linha:\n", media,"\n")
print("Matriz com a media de cada linha subtraida:\n", matriz_linha_media,"\n")
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
import numpy as np
matriz = np.random.rand(4,4)

diagonal_principal = np.diag(matriz)
diagonal_secundaria = np.diag(np.fliplr(matriz))
media_diagonal = np.mean([diagonal_principal, diagonal_secundaria])
desvio_padrao = np.std([diagonal_principal, diagonal_secundaria])

print("Matriz principal:\n", matriz, "\n")
print("Diagonal principal:\n", diagonal_principal, "\n")
print("Matriz secundaria:\n", diagonal_secundaria, "\n")
print("Media das diagonais:\n", media_diagonal, "\n")
print("Desvio padrao das diagonais:\n", desvio_padrao, "\n")
```
