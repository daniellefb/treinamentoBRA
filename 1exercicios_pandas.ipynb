### 1. Ler e exibir um arquivo CSV usando pandas

```
import numpy as np
import pandas as pd

df = spark.read.csv("/FileStore/tables/Retail.CSV", header=True, sep=';', inferSchema=True)
df = df.toPandas()
df.head()

#dados.head().style.set_table_styles([dict(selector='th', props=[('text-align', 'left')]),
                                   # dict(selector='td', props=[('text-align', 'left')])])
```

### 2. Selecionar colunas específicas de um DataFrame usando pandas

```
df[['invoice_no', 'description', 'quantity']]
```

### 3. Filtrar linhas com base em uma condição usando pandas

```
df = df[df['unit_price'] >= "5"]
df.head()
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
df_agrupado = df.groupby('description').agg(n=('invoice_no','count')).reset_index()
df_agrupado.head()
```

### 5. Usar a função max em alguma coluna do DataFrame

```
df['quantity'].max()
```
